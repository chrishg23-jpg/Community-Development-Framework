```python
import numpy as np
import matplotlib.pyplot as plt

# =========================
# Parameters
# =========================
Nx, Ny = 200, 200
dx = dy = 1.0

dt = 0.01
steps = 2000

# Base engine parameters
alpha_base = 0.6      # base anti-diffusion
beta       = 0.005    # damping
gamma      = 0.0001   # global constraint

# N-dependence
N      = 100          # try 99, 100, 101 etc.
N_res  = 100          # resonance point
sigmaN = 3.0          # width of resonance

def N_resonance_factor(N, N_res=100, sigma=3.0):
    return np.exp(-((N - N_res)**2) / (2.0 * sigma**2))

alpha = alpha_base * N_resonance_factor(N, N_res, sigmaN)

print(f"N = {N}, resonance factor = {N_resonance_factor(N, N_res, sigmaN):.3f}, alpha = {alpha:.3f}")

# =========================
# Fields
# =========================
S = np.random.uniform(-1, 1, (Nx, Ny))
M = np.zeros_like(S)   # curvature memory

# Seed asymmetry
S[60:90, 60:90] += 0.5
S[120:150, 120:150] -= 0.5

# =========================
# Laplacian operator
# =========================
def laplacian(S):
    return (
        np.roll(S, 1, axis=0) + np.roll(S, -1, axis=0) +
        np.roll(S, 1, axis=1) + np.roll(S, -1, axis=1) -
        4 * S
    )

# =========================
# Time evolution
# =========================
for step in range(steps):
    lap = laplacian(S)
    global_mean = np.mean(S)

    dSdt = alpha * lap - beta * S - gamma * global_mean
    S += dt * dSdt

    M += np.abs(lap)

# =========================
# Visualization
# =========================
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.imshow(S, cmap='viridis')
plt.title(f"Scalar Field S (N={N})")
plt.colorbar()

plt.subplot(1, 2, 2)
plt.imshow(M, cmap='inferno')
plt.title(f"Curvature Memory M (N={N})")
plt.colorbar()

plt.tight_layout()
plt.show()
```

```python
import numpy as np

Nx, Ny = 200, 200
dx = dy = 1.0

dt = 0.01
steps = 2000

alpha_base = 0.6
beta       = 0.005
gamma      = 0.0001

def N_resonance_factor(N, N_res=100, sigma=3.0):
    return np.exp(-((N - N_res)**2) / (2.0 * sigma**2))

def laplacian(S):
    return (
        np.roll(S, 1, axis=0) + np.roll(S, -1, axis=0) +
        np.roll(S, 1, axis=1) + np.roll(S, -1, axis=1) -
        4 * S
    )

def run_curvature_action(N):
    alpha = alpha_base * N_resonance_factor(N)
    S = np.random.uniform(-1, 1, (Nx, Ny))

    S[60:90, 60:90] += 0.5
    S[120:150, 120:150] -= 0.5

    total_action = 0.0

    for step in range(steps):
        lap = laplacian(S)
        global_mean = np.mean(S)

        dSdt = alpha * lap - beta * S - gamma * global_mean
        S += dt * dSdt

        # curvature-action: integrated |laplacian|
        total_action += np.sum(np.abs(lap)) * dt

    return total_action
```

```python
Ns = [95, 99, 100, 101, 105]
actions = [run_curvature_action(N) for N in Ns]
print(list(zip(Ns, actions)))
```
