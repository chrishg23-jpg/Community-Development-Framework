```python
import numpy as np
import matplotlib.pyplot as plt

# ============================================================
# ARC XV — MESOSCALE ENTANGLEMENT GRADIENT TEST
# ============================================================

# 1. Generate a synthetic void → wall density profile (1D)
N = 1000
x = np.linspace(0, 1, N)

# Void = low density, Wall = high density
# Smooth gradient from 0.1 → 1.0
density = 0.1 + 0.9 * x**2

# 2. Compute entanglement density
# In HES, entanglement density increases with matter density
rho_ent = density.copy()

# 3. Curvature compensation load C(x)
# For this mesoscale test, approximate C(x) as proportional to density
C = 1 + 4 * density

# 4. Compute update-rate map f(x)
# HES rule: f = 1 / (rho_ent * C)
f = 1.0 / (rho_ent * C)

# 5. Extract void and wall update rates
f_void = f[0]          # lowest density region
f_wall = f[-1]         # highest density region
delta_f = f_void - f_wall

# 6. Compute the Temporal Moiré redshift term
delta_z_temporal = (f_void - f_wall) / f_wall

# ============================================================
# PRINT RESULTS
# ============================================================

print("=== MESOSCALE ENTANGLEMENT GRADIENT TEST ===")
print(f"Void update rate f_void: {f_void:.6f}")
print(f"Wall update rate f_wall: {f_wall:.6f}")
print(f"Δf (void - wall): {delta_f:.6f}")
print(f"Temporal Moiré redshift term Δz_temporal: {delta_z_temporal:.6f}")

# ============================================================
# PLOTS
# ============================================================

plt.figure(figsize=(14, 10))

plt.subplot(3,1,1)
plt.plot(x, density, color='purple')
plt.title("Density Profile (Void → Wall)")
plt.ylabel("Density")

plt.subplot(3,1,2)
plt.plot(x, rho_ent, color='blue')
plt.title("Entanglement Density ρ_ent(x)")
plt.ylabel("ρ_ent")

plt.subplot(3,1,3)
plt.plot(x, f, color='red')
plt.title("Update Rate f(x)")
plt.ylabel("f")
plt.xlabel("Position (Void → Wall)")

plt.tight_layout()
plt.show()
```
